```python
import os 

lap = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])

def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
name=os.listdir("real image/")
ratio=np.array([],dtype = float)
for t in name:
    image=cv2.imread("real image/"+t)
    if image is None:
        print("Could not read input image")
        continue

    image = np.array(image)
    data0 = data.where(data["img"] == t)
    data0 = data0.dropna(how="any")

    #plt.imshow(convertToRGB(image))
    x1 = int(data0['X'].values[0])
    y1 = int(data0['Y'].values[0])
    x2 = int(data0['width'].values[0] )
    y2 = int(data0['height'].values[0] )
    print(t)
    image_copy=image.copy()
    image_copy1=image.copy()

    face_copy = image_copy[y1:y1+y2,x1:x1+x2]
    #plt.imshow(convertToRGB(face_copy))
    face_region = image_copy1[y1+int(3*y2/4):y1+int(3*y2/4)+int(y2/8),x1+int(x2/2):x1+int(x2/2)+int(x2/8)]

    laplace_face = cv2.filter2D(face_copy, -1,lap)
    laplace_face = laplace_face.astype(np.int16)
    #plt.imshow(laplace_face)

    variance_face = np.var(laplace_face)
    #print(variance_face)

    laplace_region = cv2.filter2D(face_region, -1,lap)
    laplace_region = laplace_region.astype(np.int16)

    variance_region = np.var(laplace_region)
    #print(variance_region)

    ratio = np.append(ratio,variance_face)

#ratio = np.array(ratio, dtype = float)
print(ratio)
avg = np.average(ratio)    
print(avg)
    
    
    
```
