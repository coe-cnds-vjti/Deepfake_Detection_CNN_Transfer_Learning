```python
# import required packages
import cv2
import dlib
import argparse
import time
import os 
name=os.listdir("000fakeframes/")
import pandas as pd
print(name)
#image=cv2.imread("Faces/Aamir/frame30.jpg")
import tensorflow as tf
```

```python

k=0
location = []    
final_location = []
count=0
for t in name:
    count+=1
    image=cv2.imread("000fakeframes/"+t)
    if image is None:
        print("Could not read input image")
        #location = [t,-1,-1,-1,-1]
        #final_location.append(location)
        continue

    p = "mmod_human_face_detector.dat"
    cnn_face_detector = dlib.cnn_face_detection_model_v1(p)
    faces_cnn = cnn_face_detector(image, 1)

    print(count)
# loop over detected faces
    for face in faces_cnn:
        x = face.rect.left()-20
        y = face.rect.top()-20
        w = face.rect.right() - x +20
        h = face.rect.bottom() - y +20
        location = [t,x,y,w,h]
        final_location.append(location)
        break
save1 = pd.DataFrame(final_location,columns=['img','X','Y','width','height'])
save1.to_csv('face_location_000fakevideo.csv')
```
